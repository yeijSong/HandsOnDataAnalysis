라이브러리 임포트

```python
import plotly.express as px
import pandas as pd
```

데이터 준비하기

```python
df = pd.read_csv('도로교통공단_교통사고 정보.csv', encoding='euc-kr')
df.head(3)
```

```python
df.info()
```

교통사고 시간 전처리

```python
#substr 문자열 자르기를 위해 '발생년월일시'의 데이터 타입을 문자열로 변환
df= df.astype({'발생년월일시' : 'string'})
df.info()
```

```python
# 발생년월일시(2019010100)에서 시간을 나타내는 맨 뒤 두자리('00')을 '시간'으로 분리

df['발생시간'] = df['발생년월일시'].str[8:]
df.info()
```

```python
# 발생시간의 데이터 타입을 숫자로 변경

df = df.astype({'발생시간' : 'int64'})
df.info()
```

```python
# 발생년월일시(2019010100)에서 -> 20190101(발생년월) 추출 -> 2019-01-01(날짜형태)의 데이터타입으로 변경

df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8],
                              format='%Y-%m-%d',
                              errors = 'raise') # 에러를 나타내라
```

```python
df.info()
```

2019년 지역별 교통사과 데이터 시각화

```python
fig = px.scatter(df,
                 x='발생년월일시',
                 y='발생시간',
                 color='발생지시도',
                 hover_data=['발생시간']) # 그래프에 커서를 댔을 때 무엇을 보여줄지
fig.show()
```

```python
fig = px.scatter(df,
                 x='발생년월일시',
                 y='발생시간',
                 color='발생지시도',
                 size='사망자수',
                 hover_data=['발생시간']) # 그래프에 커서를 댔을 때 무엇을 보여줄지
fig.show()
```

시간대별 교통사고 사망자 현황

```python
fig = px.bar(df,
             x='사망자수',
             y='발생시간',
             orientation='h') # 사망자수의 데이터의 크기가 커서 수평방향으로 나타내는 것이 좋다
fig.show()
```

시도별 교통사고 사망자 현황

```python
fig = px.bar(df,
             x='사망자수',
             y='발생지시도',
             orientation='h',
             title='시도별 교통사고 사망자 현황')
fig.show()
```

**대구광역시의 교통사고 발생데이터 분석**

데이터 준비하기

```python
df1=df[df.발생지시도=='대구']
df1.info()
```

교통사고 발생 시간별/지역구별 사망자 현황

```python
# 교통사고 발생 시간별/지역구별 사망자 현황을 수평방향으로 그려보기
# 조금 더 강조하고 싶은 데이터를 x축에 둔다.

fig = px.bar(df1,
             x='사망자수',
             y='발생시간',
             color='발생지시군구',
             orientation='h',
             title='대구광역시 지역구에 따른 시간대별 교통사고 사망자 현황')
fig.show()
```

지도 활용하여 교통사고 현황 분석

지도를 그릴 수 있는 라이브러리 불러오기

```python
import folium

map = folium.Map(location=[36.351, 127.386]) # 대전시청의 좌표
# 지도 내에서 기준이 되는 좌표를 제시
# 내가 표현하고 싶은 지역의 시청의 좌표를 보통 기준으로 삼는다.
map
```

지도에 필요한 정보만 담기<br>
-> 대전의 교통사고 현황을 지도에 표출

```python
df_map = df[df.발생지시도=='대전']
df_map.info()
```

사망자수, 부상자수, 중상자수, 경상자수 전처리(지도에 원을 그리기 위해)

```python
df_map=df_map.astype({'사망자수':'float64'})
df_map=df_map.astype({'부상자수':'float64'})
df_map=df_map.astype({'중상자수':'float64'})
df_map=df_map.astype({'경상자수':'float64'})
```

```python
df_map.info()
```

사망자수, 경상자수에 따른 CircleMarker 지도에 표출하기

```python
for n in df_map.index:
  cnt=df_map['사망자수'][n] + df_map['경상자수'][n] # 사망자수 + 경상자수 -> CircleMarker 사이즈
  folium.CircleMarker([df_map['위도'][n],df_map['경도'][n]],
                      radius = cnt*10, # 반지름 크기 설정, 좀 크게 보기 위해 10을 곱함
                      popup = df_map['사고유형'][n],
                      color = 'red',
                      fill_color = 'orange').add_to(map)

map.save('대전교통사고현황.html')
```
